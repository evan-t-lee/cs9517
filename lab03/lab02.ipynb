```python
import cv2
from matplotlib import pyplot as plt
from scipy import ndimage

img = cv2.imread("syd_house.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

sift = cv2.xfeatures2d.SIFT_create(nfeatures=250, contrastThreshold=0.06)
kp = sift.detect(gray, None)

img = cv2.drawKeypoints(gray, kp, img)

plt.imshow(img, cmap='gray', vmin=0, vmax=255)

cv2.imwrite('Task_1.png', img)
plt.show()

# --------------------------- 

rotated = ndimage.rotate(img, 255)
gray = cv2.cvtColor(rotated, cv2.COLOR_BGR2GRAY)

sift = cv2.xfeatures2d.SIFT_create(nfeatures=250, contrastThreshold=0.06)
kp = sift.detect(gray, None)

rotated = cv2.drawKeypoints(gray, kp, rotated)

plt.imshow(rotated, cmap='gray', vmin=0, vmax=255)

# not part of submission
cv2.imwrite('Task_2.png', rotated)
plt.show()

# ---------------------------

sift = cv2.xfeatures2d.SIFT_create(nfeatures=250, contrastThreshold=0.06)

kp1, des1 = sift.detectAndCompute(img, None)
kp2, des2 = sift.detectAndCompute(img, None)

bf = cv2.BFMatcher()
matches = bf.knnMatch(des1,des2,k=2)

# Ratio test
good = []
for m, n in matches:
    if m.distance < 0.75*n.distance:
        good.append([m])
        
# cv.drawMatchesKnn expects list of lists as matches.
img_match = cv2.drawMatchesKnn(img, kp1, img, kp2, good, None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

plt.imshow(img_match, cmap='gray', vmin=0, vmax=255)

cv2.imwrite('Task3_0.png', img_match)
plt.show()

# ---------------------------

rotated_60 = ndimage.rotate(img, 60)

sift = cv2.xfeatures2d.SIFT_create(nfeatures=250, contrastThreshold=0.06)

kp1, des1 = sift.detectAndCompute(img, None)
kp2, des2 = sift.detectAndCompute(rotated_60, None)

bf = cv2.BFMatcher()
matches = bf.knnMatch(des1,des2,k=2)

# Ratio test
good = []
for m,n in matches:
    if m.distance < 0.75*n.distance:
        good.append([m])
        
# cv.drawMatchesKnn expects list of lists as matches.
img_match = cv2.drawMatchesKnn(img, kp1, rotated_60, kp2, good, None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

plt.imshow(img_match, cmap='gray', vmin=0, vmax=255)

cv2.imwrite('Task3_60.png', img_match)
plt.show()

# ---------------------------

rotated_120 = ndimage.rotate(img, 120)

sift = cv2.xfeatures2d.SIFT_create(nfeatures=250, contrastThreshold=0.06)

kp2, des2 = sift.detectAndCompute(rotated_120, None)

bf = cv2.BFMatcher()
matches = bf.knnMatch(des1,des2,k=2)

# Ratio test
good = []
for m,n in matches:
    if m.distance < 0.75*n.distance:
        good.append([m])
        
# cv.drawMatchesKnn expects list of lists as matches.
img_match = cv2.drawMatchesKnn(img, kp1, rotated_120, kp2, good, None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

plt.imshow(img_match, cmap='gray', vmin=0, vmax=255)

cv2.imwrite('Task3_120.png', img_match)
plt.show()
```
